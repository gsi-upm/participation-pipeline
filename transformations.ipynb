```python
from elasticsearch import Elasticsearch
import json
import requests

# create a client instance of the library
elastic_client = Elasticsearch()

query = {
    'size' : 10000,
    'query': {
        'match_all' : {}
    }
}

response = elastic_client.search( index='participation', body=query)
elastic_docs = response["hits"]["hits"]

print(json.dumps(elastic_docs[:100], indent=3))
```

```python
import numpy as np
import pandas as pd
import json

file = open('backup_demos.json', 'r')
Lines = file.readlines()

elastic_docs = []
for line in Lines:
    elastic_docs.append(json.loads(line.strip()))
    
# for doc in elastic_docs:
#     if doc["_index"] != "participation":
#         elastic_docs.remove(doc)
    
print(json.dumps(elastic_docs[:100], indent=3))
```

```python
import requests

host = 'http://localhost:8000/api/'
tweet_list = []

for doc in elastic_docs:
    if doc["_index"] != "participation":
        continue

    tweet = doc["_source"]
    
    if tweet["schema:inLanguage"] == "en":
        r = requests.post(host, data={'i':tweet["schema:articleBody"], 'apiKey': "", 'algo': 'mft'})
        tweet["mft:result"] = r.json()["entries"][0]["mft:result"]
        
        try:
            for k,v in tweet["mft:result"].items():
                tweet["radar"].append({"key": k, "value": v*100/tweet["word_count"]})
                tweet["radar_mft"].append({"key": k, "value": v*100/tweet["word_count"]})
        except KeyError:
            for k,v in tweet["mft:result"].items():
                tweet["radar"] = [{"key": k, "value": v*100/tweet["word_count"]}]
                tweet["radar_mft"] = [{"key": k, "value": v*100/tweet["word_count"]}]
                
        doc["_source"] = tweet
        tweet_list.append(doc)
```

```python
print(json.dumps(tweet_list[:100], indent=3))
```

```python
with open('mft.json', 'w') as mft_dumped:
    for doc in tweet_list:
        json.dump(doc, mft_dumped)
        mft_dumped.write('\n')
```

```python
import numpy as np
import pandas as pd
import json

file = open('alternative.json', 'r')
Lines = file.readlines()

elastic_docs = []
for line in Lines:
    elastic_docs.append(json.loads(line.strip()))
    
print(json.dumps(elastic_docs[:100], indent=3))

words = {
    "HarmVirtue": "Harm Virtue",
    "HarmVice": "Harm Vice",
    "FairnessVirtue": "Fairness Virtue",
    "FairnessVice": "Fairness Vice",
    "IngroupVirtue": "Ingroup Virtue",
    "IngroupVice": "Ingroup Vice",
    "AuthorityVirtue": "Authority Virtue",
    "AuthorityVice": "Authority Vice",
    "PurityVirtue": "Purity Virtue",
    "PurityVice": "Purity Vice"
}

tweet_list = []
for doc in elastic_docs:
    tweet = doc["_source"]
    
    keys = list(tweet["mft:result"].keys())
    for key in keys:
        if key == "MoralityGeneral":
            del tweet["mft:result"][key]
        else:
            tweet["mft:result"][words[key]] = tweet["mft:result"][key]
            del tweet["mft:result"][key]
        
#     for radar in tweet["radar"]:
#         if radar["key"] in words.keys():
#             tweet["radar"].append({
#                 "key": words[radar["key"]],
#                 "value": radar["value"]
#             })
#             tweet["radar"].remove(radar)
            
#         elif radar["key"] == "MoralityGeneral":
#             tweet["radar"].remove(radar)
            
    try:
        for radar in tweet["radar_mft"]:
            if radar["key"] in words.keys():
                tweet["radar_mft"].append({
                    "key": words[radar["key"]],
                    "value": radar["value"]
                })
                tweet["radar_mft"].remove(radar)

            elif radar["key"] == "MoralityGeneral":
                tweet["radar_mft"].remove(radar)
    except:
        pass

    doc["_source"] = tweet
    tweet_list.append(doc)
            
    
print(json.dumps(tweet_list[:100], indent=3))
```

```python
with open('treemap.json', 'w') as mft_dumped:
    for doc in tweet_list:
        json.dump(doc, mft_dumped)
        mft_dumped.write('\n')
```

```python
import numpy as np
import pandas as pd
import json

file = open('participation_mft.json', 'r')
Lines = file.readlines()

elastic_docs = []
for line in Lines:
    elastic_docs.append(json.loads(line.strip()))

tweet_list = []
for doc in elastic_docs:
    tweet = doc["_source"]
    
    if "MoralityGeneral" in tweet["mft:result"].keys():
        del tweet["mft:result"]["MoralityGeneral"]
       
    for radar in tweet["radar"]:            
        if radar["key"] == "MoralityGeneral":
            tweet["radar"].remove(radar)
            
    try:
        for radar in tweet["radar_mft"]:
            if radar["key"] == "MoralityGeneral":
                tweet["radar_mft"].remove(radar)
    except:
        pass

    doc["_source"] = tweet
    tweet_list.append(doc)
            
    
print(json.dumps(tweet_list[:100], indent=3))

with open('treemap2.json', 'w') as mft_dumped:
    for doc in tweet_list:
        json.dump(doc, mft_dumped)
        mft_dumped.write('\n')
```
