```python
import twint
import nest_asyncio
nest_asyncio.apply()

t = twint.Config()

query = "#9N OR #9N2014 OR #Proces OR #Donbas OR #Catalunya OR #VoteLeave OR #SpainIsAFascistState OR #RepublicaCatalana OR #ANC"

t.Search = query
t.Store_object = True
t.Limit = 1000
t.Near = "europe"
t.Profile_full = True

twint.run.Search(t)

tweet_list = twint.output.tweets_list
places = list(filter(lambda x: x.place, tweet_list))
places
```

```python
print(len(places))
print(len(tweet_list))
print(len(places)*100/len(tweet_list))

for tweet in places:
    print(tweet.place)
    print(tweet.datetime)
```

```python
languages = {}
for tweet in tweet_list:
    try:
        languages[tweet.lang] += 1
    except KeyError:
        languages[tweet.lang] = 1
        
languages
```

```python
c = twint.Config()
users = set()
locations = {}

for tweet in tweet_list:
    username = tweet.username
    if username in users:
        print(username + ': ' + locations[username])
        continue
    
    c.Username = tweet.username
    c.Store_object = True
    
    try:
        twint.run.Lookup(c)
    except RefreshTokenException:
        continue
    
    users.add(username)
    locations.update({username: twint.output.users_list[-1].location})
    
    print(username + ': ' + locations[username])
    
print(len(tweet_list))
print(len(users))
print(len(locations))
```

```python
import time
import json
from rdflib import Graph, URIRef, Namespace

sioc = Namespace('http://rdfs.org/sioc/ns#')
sioc_t = Namespace('http://rdfs.org/sioc/types#')
dc = Namespace('http://purl.org/dc/terms/')
schema = Namespace('http://schema.org/')
narr = Namespace('http://www.gsi.upm.es/ontologies/participation/narrative/ns#')
sliwc = Namespace('http://www.gsi.upm.es/ontologies/participation/sliwc/ns#')
moral = Namespace('http://www.gsi.upm.es/ontologies/participation/moral/ns#')
dbpedia = Namespace('https://www.dbpedia.org/')
nif = Namespace('http://persistence.uni-leipzig.org/nlp2rdf/ontologies/nif-core#')
rdfs = Namespace('http://www.w3.org/2000/01/rdf-schema#')

context = {
    "id": sioc.id,
    "content": sioc.content,
    "created": dc.created,
    "has_creator": sioc.has_creator,
    "name": sioc.name,
    "mentions": schema.mentions,
    "hasIdeology": narr.hasIdeology,
    "hasAnnotation": sliwc.hasAnnotation,
    "hasCategory": sliwc.hasCategory,
    "narr_ratio": narr.ratio,
    "ratio": sliwc.ratio,
    "label": rdfs.label,
    "isString": nif.isString,
    "locationCreated": schema.locationCreated,
    "inLanguage": schema.inLanguage
}

file = open('treemap2.json', 'r')
Lines = file.readlines()

elastic_docs = []
for line in Lines:
    elastic_docs.append(json.loads(line.strip()))
    
# print(json.dumps(elastic_docs[:10], indent=3))
tweet_list = []
graphs = []
n=0
for doc in elastic_docs:
    try:
        g = graphs[int(n/100)]
    except IndexError:
        graphs.append(Graph())
        g = graphs[int(n/100)]
    
    n+=1
    
    tweet = doc["_source"]
#     print(json.dumps(tweet, indent=3))

    # tweet = {}

    tweet["@context"] = context
    tweet["@type"] = [sioc.Post]
    tweet["id"] = int(tweet["@id"].split("/")[-1])

    tweet["content"] = tweet["schema:articleBody"]
    tweet["created"] = tweet["schema:datePublished"]
    tweet["has_creator"] = {
        "@type": sioc.UserAccount,
        "id": tweet["schema:creator"],
        "name": tweet["schema:creator"]
    }

    tweet["mentions"] = list(map(
        lambda ht: {"@type": sioc_t.Tag, "label": ht},
        tweet["schema:keywords"]
    ))

#     tweet["schema:locationCreated"] = tweet["schema:locationCreated"]
#     tweet["schema:inLanguage"] = tweet["schema:inLanguage"]

    # tweet["schema:mentions"].extend(list(map(
    #     lambda mention: {

    #     },
    #     tweet.mentions
    # )))

    # tweet["schema:interactionStatistic"] = [
    #     {
    #         "@type": "schema:InteractionCounter",
    #         "schema:userInteractionCount": tweet.likes_count,
    #         "schema:interactionType": "schema:LikeAction"
    #     },
    #     {
    #         "@type": "schema:InteractionCounter",
    #         "schema:userInteractionCount": tweet.retweets_count,
    #         "schema:interactionType": "schema:ShareAction"
    #     }
    # ]
    
    tweet["hasAnnotation"] = []
    
    for ideology in tweet["ideology"]:
        tweet.update({"hasIdeology": {"@type": narr[ideology.replace(" ", "_")]}})
        tweet["hasAnnotation"].append({
            "@type": [sliwc.AggregatedAnnotation, narr.Annotation],
            "hasCategory": {"@type": narr[tweet[ideology]+ideology.replace(" ", "_")]},
            "narr_ratio": 1.0
        })

    tweet["hasAnnotation"].extend(list(map(
        lambda x: {
            "@type": sliwc.AggregatedAnnotation,
            "hasCategory": {"@type": sliwc[x["key"]]},
            "ratio": x["value"]/100
        }, tweet["radar_social"]
    )))
    
    tweet["hasAnnotation"].extend(list(map(
        lambda x: {
            "@type": sliwc.AggregatedAnnotation,
            "hasCategory": {"@type": sliwc[x["key"]]},
            "ratio": x["value"]/100
        }, tweet["radar_concerns"]
    )))
    
    tweet["hasAnnotation"].extend(list(map(
        lambda x: {
            "@type": sliwc.AggregatedAnnotation,
            "hasCategory": {"@type": sliwc[x["key"]]},
            "ratio": x["value"]/100
        }, tweet["radar_affective"]
    )))
    
    tweet["hasAnnotation"].extend(list(map(
        lambda x: {
            "@type": sliwc.AggregatedAnnotation,
            "hasCategory": {"@type": sliwc[x["key"]]},
            "ratio": x["value"]/100
        }, tweet["radar_drives"]
    )))
    
    if "radar_mft" in tweet:
        tweet["hasAnnotation"].extend(list(map(
            lambda x: {
                "@type": [sliwc.AggregatedAnnotation, moral.Annotation],
                "hasCategory": {"@type": moral[x["key"].replace(" ", "")]},
                "ratio": x["value"]/100
            },tweet["radar_mft"]
        )))
        tweet.pop("radar_mft")
        
    tweet.pop("radar")
    tweet.pop("radar_drives")
    tweet.pop("radar_concerns")
    tweet.pop("radar_affective")
    tweet.pop("radar_social")
    tweet.pop("word_count")
    tweet.pop("liwc:result")
    for ideology in tweet.pop("ideology"):
        tweet.pop(ideology)
    tweet.pop("narrative")
    tweet.pop("counter-narrative")
    tweet.pop("alternative-narrative")
    tweet.pop("mft:result")
    
    tweet.pop("onyx:hasEmotionSet")
    tweet.pop("marl:hasOpinion")
    tweet.pop("schema:keywords")
    tweet.pop("schema:creator")
    tweet.pop("schema:datePublished")
    tweet.pop("schema:articleBody")
    tweet.pop("schema:about")
    tweet.pop("schema:headline")
    tweet.pop("schema:author")
    tweet.pop("schema:search")
    
    tweet["isString"] = tweet.pop("nif:isString")
    if 'schema:locationCreated' in tweet:
        tweet["locationCreated"] = tweet.pop("schema:locationCreated")
    tweet["inLanguage"] = tweet.pop("schema:inLanguage")
    
    tweet_list.append(tweet)
    g.parse(data=json.dumps(tweet), format='json-ld')

# print(json.dumps(tweet_list[:10], indent=3))

# print(g.serialize(format='turtle', indent=3).decode('utf-8'))
n = 0
for g in graphs:
    g.serialize(format='turtle', destination='participation{}.ttl'.format(n))
    n+=1
```
